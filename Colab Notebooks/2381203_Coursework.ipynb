```python
import numpy as np
import skimage.feature
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Load the data
X = np.load('/content/drive/MyDrive/Colab Notebooks/DS_Xdata.npy')  # low resolution images
Y = np.load('/content/drive/MyDrive/Colab Notebooks/Ydata.npy')     # the labels

# Shuffle the data
X, Y = shuffle(X, Y, random_state=42)

print("Data loaded and shuffled.")
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
# First, split off the test set (10% of the original data)
X_intermediate, X_test, Y_intermediate, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)

# This results in 80% train and 10% val of the original data
X_train, X_val, Y_train, Y_val = train_test_split(X_intermediate, Y_intermediate, test_size=0.1111, random_state=42)  # approximately 0.1111 â‰ˆ 1/9

print("Data split into training, validation, and testing sets.")
print("Training set shape:", X_train.shape)
print("Validation set shape:", X_val.shape)
print("Testing set shape:", X_test.shape)
```

```python
# Initialize the CNN
model = Sequential()
# Add the convolutional base
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(61, 61, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
# Add the fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(21, activation='softmax'))  # 21 classes corresponding to the dataset

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


# Train the model
history = model.fit(X_train, Y_train, epochs=10, validation_data=(X_val, Y_val), batch_size=32)

print("Model has been trained.")
```

```python
# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, Y_test)
print("Test accuracy:", test_acc)
```
